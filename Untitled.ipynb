```python
clips = sorted(glob.glob("results/3D/iou0.01/data/*"))

for clip in clips[6:]:
    print(clip)
    clip_dict = {}
    with open(clip, 'r') as f:
        lines = f.readlines()
    for line in lines:
        label = line.strip().split(' ')
        frame_id = int(label[0])
        track_id = int(label[1])
        animal_type = label[2]
        truncation = float(label[3])
        occlusion = float(label[4])  # 0:fully visible 1:partly occluded 2:largely occluded 3:unknown
        alpha = float(label[5])
        box2d = list(np.array((float(label[6]), float(label[7]), float(label[8]), float(label[9])), dtype=np.float32))
        l = float(label[10])
        h = float(label[11])
        w = float(label[12])
        loc = list(np.array((float(label[13]), float(label[14]), float(label[15])), dtype=np.float32))
        dis_to_cam = np.linalg.norm(loc)
        ry = float(label[16])
        score = float(label[17]) if label.__len__() == 18 else -1.0
    
        if track_id not in clip_dict.keys(): 
            clip_dict[track_id] =  {'type': animal_type,
                                     'box2d': [box2d],
                                     'size': [[h,w,l]],
                                     'pos': [loc],
                                     'angle': [ry],
                                     'score': [score],
                                     'frame_id': [frame_id],
                                     'dis_to_cam': [dis_to_cam]
                                    }
        else:
            clip_dict[track_id]['box2d'].append(box2d)
            clip_dict[track_id]['size'].append([h,w,l])
            clip_dict[track_id]['pos'].append(loc)
            clip_dict[track_id]['angle'].append(ry)
            clip_dict[track_id]['score'].append(score)
            clip_dict[track_id]['frame_id'].append(frame_id)
            clip_dict[track_id]['dis_to_cam'].append(dis_to_cam)
    break
clip_dict.keys()
```

```python
def ema(values, period=5):
    values = np.array(values)
    df = pd.DataFrame(values)
    return df.ewm(alpha=0.1).mean()
```

```python
#print(clip_dict[3]['frame_id'])
for k,track in clip_dict.items():
    if len(track['pos']) > 15:
        vel = []
        dif = []
        for i,pos in enumerate(track['pos'][:-1]):
            if track['frame_id'][i]+1 == track['frame_id'][i+1]:
                diff = np.array(track['pos'][i]) - np.array(track['pos'][i+1])
                #if diff[0] < 0.05: diff[0]=0.
                #if diff[1] < 0.05: diff[1]=0.
                #if diff[2] < 0.05: diff[2]=0.
                dif.append(diff)
        track['id'] = k   
        track['diff'] = np.array(ema(dif))
        track['dist'] = np.sqrt(track['diff'][:,0]**2+track['diff'][:,1]**2)
        track['vel'] = (track['dist']/(1/15))*3.6
        track['avg_vel'] = [np.mean(track['vel']),np.std(track['vel'])]
        track['avg_size'] = [np.mean(track['size'], axis=0),np.std(track['size'], axis=0)]
        track['avg_score'] = [np.mean(track['score'], axis=0),np.std(track['score'], axis=0)]
        track['dis_to_cam'] = [np.mean(track['dis_to_cam'], axis=0),np.std(track['dis_to_cam'], axis=0)]
```

```python
from kitti_utils import project_to_image

def reproj(pcl,pts_3d,calib):
    R = np.array([[1, 0, 0],
                      [0, 0, 1],
                      [0,-1, 0]])
    T = np.array(calib.extrinsics).reshape(4,4)
    box_pts = o3d.geometry.PointCloud()
    box_pts.points = o3d.utility.Vector3dVector(pts_3d.T)
    box_pts.rotate(R.T, np.array(pcl.get_center()))
    box_pts.transform(T)

    proj_pcl,_ = cv2.projectPoints(np.array(box_pts.points),
                                        np.array([0,0,0], np.float),
                                        np.array([0,0,0], np.float),calib.P[:3,:3],None)
    proj_pcl = proj_pcl.squeeze(1)
    return proj_pcl

result_dir = os.path.join(result_root, '%s' % (seq))

objs = []
for i in range(80,150):
    img_fn = images[i]
    result_tmp = os.path.join(result_dir, img_fn.replace(".jpg",".txt"))
    object_res = read_label(result_tmp)
    
    objs.append(object_res)

out_dict = dict()
for obj in objs:
    for o in obj:
        p = [o.t[0],o.t[1],o.t[2]]
        if o.id in out_dict.keys():
            out_dict[o.id].append(p)
        else:
            out_dict[o.id] = []
            out_dict[o.id].append(p)

        
print(out_dict)
id_list = [1,3,4,5]
cs = ["orange","red","purple", "green"]
id_to_color = {1: "orange", 3: "red", 4: "purple", 5: "green"}
for color_idx,id in enumerate(id_list):

    color_tmp = color_map_color(id, vmax=max_color)
    box3d_pts_2d = reproj(pcl,np.array(out_dict[id]).T, calib)
    box3d_pts_2d += b
    
    ax.plot(box3d_pts_2d[::5,0],box3d_pts_2d[::5,1], color=id_to_color[id], linewidth=2)
plt.savefig("tracklets1.png", bbox_inches='tight', dpi=200)
plt.show()
```

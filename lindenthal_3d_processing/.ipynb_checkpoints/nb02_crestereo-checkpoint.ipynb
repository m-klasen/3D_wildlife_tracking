```python
from crestereo import CREStereo, CameraConfig

import os 
import glob
import numpy as np
from PIL import Image

os.environ["OPENCV_IO_ENABLE_OPENEXR"]="1"
import cv2

import matplotlib.pyplot as plt 
from pathlib import Path

```

```python
crestereo = CREStereo(
        f'weights/crestereo/crestereo_combined_iter5_720x1280.onnx',
        camera_config=CameraConfig(0.25, 1591.4101, ) ,
        max_dist=100,
    )
```

```python
#DATA_DIR= "/media/mlk/storage/pltsdrf/detections_images/20220308132707"
DATA_DIR= "/media/mlk/storage/data/train/pltsdorf/20220213103546"
baseline = 0.25
fx = 1591.4101
```

```python
all_samples = sorted(glob.glob(DATA_DIR + '/left_colorized/*.png'))

num_samples = len(all_samples)
print('=> %d samples found in the data dir' % num_samples)
os.makedirs(os.path.join(DATA_DIR, "disp_crestereo"), exist_ok=True)

for i, sample_name in enumerate(all_samples):
    if i % 100 == 0:
        print('=> Inferencing %d/%d' % (i, num_samples))

    left_name = sample_name
    right_name = left_name.replace('left', 'right')
    disp_name = os.path.join(DATA_DIR, "disp_crestereo")

    left = np.array(Image.open(left_name).convert('RGB')).astype(np.float32)
    right = np.array(Image.open(right_name).convert('RGB')).astype(np.float32)
    #left = left[200:,:720]
    #right = right[200:,:720]


    disp_crestereo = crestereo(left, right)
    disp_crestereo = cv2.resize(disp_crestereo, (left.shape[1], left.shape[0]), interpolation=cv2.INTER_NEAREST)

    cv2.imwrite(os.path.join(disp_name, f"{Path(left_name).name[:-4]}.exr"), disp_crestereo)
```

```python
plt.figure(0,figsize=(15,15))
disparity = cv2.imread("/media/mlk/storage/data/valid/pltsdorf/20220209151344/disp_aanet/000091_pred.png", -1)/256.

depth  = baseline*fx/disparity
depth[depth>40] = 0.
d = plt.imshow(depth, "plasma")
plt.colorbar(d,fraction=0.026, pad=0.04)
plt.show()


plt.figure(0,figsize=(15,15))
disparity = cv2.imread(DATA_DIR+"/disp_crestereo/000091.exr", -1)*2

depth  = baseline*fx/disparity
depth[depth>40] = 0.
d = plt.imshow(depth, "plasma")
plt.colorbar(d,fraction=0.026, pad=0.04)
plt.show()
```

```python
os.makedirs(DATA_DIR+"/depth", exist_ok=True)
for disp_file in sorted(glob.glob(DATA_DIR+"/disp_crestereo/*")):
    depth_file = disp_file.replace("/disp_crestereo","/depth").replace("_pred.png",".exr")
    disparity = cv2.imread(disp_file, -1)/256.
    depth = np.zeros_like(disparity).astype(np.float32)

    if baseline==0.25:
        depth = baseline*fx/disparity
        depth[depth>40] = 0.
    else:
        depth  = baseline*fx/disparity
        depth[depth>40] = 0.
    
    
    cv2.imwrite(depth_file,depth.astype(np.float32))
```

```python
baseline = 0.05
fx =  424.74
DATA_DIR= "/media/mlk/storage/AMMOD/ammod_realsense/data/2d/20201219163632"

all_samples = sorted(glob.glob(DATA_DIR + '/infrared/*left.jpg'))

num_samples = len(all_samples)
print('=> %d samples found in the data dir' % num_samples)
os.makedirs(os.path.join(DATA_DIR, "disp_crestereo"), exist_ok=True)

for i, sample_name in enumerate(all_samples):
    if i % 100 == 0:
        print('=> Inferencing %d/%d' % (i, num_samples))

    left_name = sample_name
    right_name = left_name.replace('left', 'right')
    print(left_name,right_name)
    disp_name = os.path.join(DATA_DIR, "disp_crestereo")

    left = np.array(Image.open(left_name).convert('RGB')).astype(np.float32)
    right = np.array(Image.open(right_name).convert('RGB')).astype(np.float32)
    left = left[100:,:720]
    right = right[100:,:720]


    disp_crestereo = crestereo(left, right)
    disp_crestereo = cv2.resize(disp_crestereo, (left.shape[1], left.shape[0]), interpolation=cv2.INTER_NEAREST)

    cv2.imwrite(os.path.join(disp_name, f"{Path(left_name).name[:-4]}.exr"), disp_crestereo)
```
